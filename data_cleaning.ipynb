```python
import pandas as pd
df=pd.read_excel("./dataset/Customer Call List.xlsx")
df1=df.copy()
```

```python
df
```

```python
'''
1. the data set has many NaN values in it 
2. the phone numbers are in different format 
3. the last names have some unuseful characters in it
4. the data set has a non_useful column that has to be removed
5. the addresses are not standard
6. in the donot contact column, if the value is "yes" , it needs to be removed
   from the dataset
7. "yes" and "no" are given in different format, they need to be standardized
8. dataset has duplicate values in it
'''
```

```python
# the first step of the data cleaning process is to remove the duplicate values
df.drop_duplicates(inplace=True)
df
```

```python
#next we drop the irrelevant column 
df.drop(columns="Not_Useful_Column", inplace=True)
df
```

```python
#now we move from column to column and check what needs fixing
#targeting the last name column
#the last name column has "special characters like '.', '/', '_' "
#this can be removed using the strip function

#strip() function removes the "specified characters" from either the "extreme left"
#or the "extreme right" of the string

#lstrip():removes from extreme left
#rstrip():removes from extreme right
#strip(): in general focuses on removing from both extreme left and right
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")

df
```

```python
# all of them can be removed using a single strip command at once as well
df["Last_Name"] = df["Last_Name"].str.strip("./_")
df
```

```python
#now targeting the phone number columns
#first we will converts all the phone numbers to be just digits 
#then we will convert them as 123-456-7890
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(r"\D","",regex=True)
df
#here anything that is not a digit will be replaced by blank space
#so "special characters","letters" will be replaced by blank space


#df["Phone_Number"]=df["Phone_Number"].str.replace(r"[^0-9A-Za-z]","",regex=True)
#here anything that is not a number, or a character(upper and lower case) will be replaced by 
#blank space



#"r" is used to treat "\D" as a raw text
#"\D" means, anything that is not a digit
#regex= regular expression = it helps to search and replace characters easily
#"\d": any digit between 0 to 9
#"\s": any space
#"\w": any letter or number
#"[A-Z]": any capital letter
#"[a-z]": any small letter
#"[0-9]": any number
#"[^0-9]": any thing that is not a digit from 0-9
#"[^0-9A-Za-z]":any thing that is not, digit,capital letter, small letter
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace(r"^\s*$","NaN",regex=True)
df
```

```python
#now we conver the number to our desired format
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace("NaN--", "Missing")

df["Phone_Number"]=df["Phone_Number"].str.replace("nan--", "Missing")

df
```

```python
'''
what we did in the above 2 cells 
A. UISNG THE LAMBA FUNCTION
we wanted to apply slicing so we first converted the numbers into srting 
we uses .apply(lambda x:str(x))
this is a single line function that stores each values of phone number column into x , and returns
str(x)

then on the string we applied slicing , by using the lambda function again 
.apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:10])

this adds "-" between the string 

B. replacing nan values
we replaced the nan values with missing using .str.replace("nan","missing")
'''
```

```python
#next targeting address 
# we split the address into street, state, and pincode for better readability 
df[["Street","State","PIN"]]=df["Address"].str.split(",", n=2, expand=True)
df=df[["CustomerID","First_Name","Last_Name","Phone_Number","Address","Street","State","PIN","Paying Customer","Do_Not_Contact"]]
df
```

```python
#deleting the address column for better readabiltiy 
df=df.drop(columns="Address")
df["Street"]=df["Street"].str.replace("N/a","None")
df
```

```python
#targeting paying customer and donot contact 
#convert "Yes" to "Y" and "No" to "N"

df["Paying Customer"]= df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"]= df["Paying Customer"].str.replace("No","N")
df["Paying Customer"]= df["Paying Customer"].str.replace("N/a","None")

df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("No","N")
df["Do_Not_Contact"]=df["Do_Not_Contact"].fillna("None")

df
```

```python
#now we have targeted all the individual columns and standardized the values
#now we have to delete the rows which have said "donot_contact=Y"

for i in df.index:
    if(df.loc[i,"Do_Not_Contact"]=="Y"):
        df.drop(i,inplace=True)

df
#now we have all the customers that can be contacted
```

```python
#we will also convert the none values inside donot conatct to "N"
#because these customers can be contacted

df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("None","N")
df
```

```python
#next we will delete the customers who have not provided a phone number
for i in df.index:
    if(df.loc[i,"Phone_Number"]=="Missing"):
        df.drop(i,inplace=True)

df
```

```python
#lastly we will reset the index , so the person doesnot think that some values are missing

#df.reset_index(), this resets the index, but saves the previous one into a new column

df.reset_index(drop=True, inplace=True)
df
```

```python
#finally we have our cleaned data 
#below is the old messy data and above is the cleaned and standard data data 

old_data=pd.read_excel("./dataset/Customer Call List.xlsx")
old_data
df.to_excel("./dataset/cleaned Customer Call List.xlsx",index=False)
```
